```python
import pandas as pd
import numpy as np
import seaborn

from scipy import stats
from matplotlib import pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.graphics.gofplots import ProbPlot
```

# 예제 1-1.

```python
handspan = pd.read_table("handspan.txt", sep="\t")
seaborn.set(rc={'figure.figsize' : (3,3)})
seaborn.relplot(x = 'HandSpan', y = "Height", data=handspan)
plt.show()
```

HandSpan을 x축으로, Height를 y축으로 하여 두 변수에 대한 산점도를 그려보면 다음과 같다.  
산점도의 분포 형태로 보아 두 변수 간에 양의 상관관계가 있음을 관차할 수 있다.


# 예제 1-2.

```python
model = ols("Height ~ HandSpan", handspan).fit()
intercept = model.params[0]
coefficient = model.params[1]
print(model.summary())
print("Height = %s * HandSpan + %s" %(coefficient, intercept))
```

독립변수 x = HandSpan, 종속변수 y = Height로 두고 단순 선형 회귀 분석을 시행한 결과이다.  

회귀모형의 유의성 검정 결과를 살펴보면, 검정 통계량 (F-statistic)은 199.2이고 유의확률 (Prob(F-statistic))은 0에 근사함을 알 수 있다.  
따라서 유의수준 5%에서 회귀직선은 유의하다고 볼 수 있다.  
또한 결정계수 (Multiple R-squared)는 0.547으로써 전체 자료의 산포 중 약 54.7%가 회귀직선으로 설명이 가능한 것을 알 수 있다.

적합된 회귀직선은 Height = 1.56 * HandSpan + 35.525이다. 
1.56로 표현되는 회귀직선의 기울기의 통계적 유의성을 검정한 결과, t-통계량이 14.113이고, 유의확률이 0에 근사하므로 유의하다.  
35.525로 표현되는 회귀직선의 절편의 통계적 유의성을 검정한 결과, t-통계량이 15.339이고, 유의확률이 0에 근사하므로 유의하다.  
따라서, 유의수준 5%에서 이 회귀직선은 유의하다고 말할 수 있다.  

# 예제 1-3.

```python
x = np.arange(15, 30)
y = intercept + coefficient * x

plt.figure(figsize = (7, 7))
plt.scatter(handspan.HandSpan, handspan.Height)
plt.plot(x, y)
plt.xlabel("HandSpan")
plt.ylabel("Height")
plt.show()
```

정의역 x (HandSpan)의 구간을 [15, 30]으로 하고 두 변수 (x=HandSpan, y=Height)의 산점도와 둘 사이의 회귀직선을 함께 나타내었다.  
대체로 회귀직선이 분포한 점들의 중간을 통과하는 것을 알 수 있다.

# 예제 1-4.

```python
# 1. 잔차도 (residual plot) : Residuals vs. Fitted
f = model.fittedvalues # 예측값 
r = model.resid # 잔차
ar = np.abs(r)

ar_sort = ar.sort_values(ascending = False)
top_10_ar = ar_sort[0:10]

seaborn.set(rc={'figure.figsize':(4,4)})
seaborn.residplot(f, r, data = handspan)

plt.title("Residuals vs Fitted")
plt.ylabel("Residuals")
plt.xlabel("Fittd values")

for i in top_10_ar.index:
    plt.annotate(i, xy = (f[i], r[i]))
plt.show()

# 2. 정규 분위수 (Normal Q-Q plot)
QQ = ProbPlot(r)
plot = QQ.qqplot(line='s', color='C0', lw=1)
plt.title("Normal Q-Q Plot")
plt.show()
```

회귀분석의 결과에 대한 잔차도 (residual plot)와 잔차의 정규 분위수 그래프를 나타내었다.  

단순선형회귀모형에서 잔차에 대해 전제하고 있는 주요 가정 중 선형성, 정규성, 등분산성, 독립성을 살펴보도록 하자.  

먼저, 잔차도를 보았을 때,
선형성 : 잔차 = 0인 직선을 중심으로 대칭적으로 고르게 분포  
등분산성 : 우측으로 갈 수록 잔차의 산포가 커지지 않고 고르게 분포  
독립성 : 별도의 유의미한 패턴이 발견되지 않음  

또한, 정규 분위수 그래프를 보았을 때,
정규성 : 대부분의 잔차는 직선 주위에 몰려있음  

비록, 잔차도에서 절댓값이 가장 큰 관측값 10개를 뽑아 그래프에 표시했을 때, 범위를 벗어난 큰 값을 갖는 것으로 확인되고,  
정규 분위수 그래프에서도 양 끝에서 직선을 벗어난 점들이 관찰되는 부분도 있긴하지만,  
전체적으로는 잔차가 선형성, 정규성, 등분산성, 독립성의 가정을 충족하므로  
단순선형회귀모형의 적용이 타당함을 알 수 있다.  

# 예제 2-1.

```python
wine = pd.read_csv('wine.csv')

y = wine["pH"]
x = wine["fixedacidity"]


model = ols("pH ~ fixedacidity", wine).fit()
intercept = model.params[0]
coefficient = model.params[1]
print(model.summary())
print("pH = %s * fixedacidity + %s" %(coefficient, intercept))
print("검정통계량 : %s, 유의확률: %s " %(model.fvalue, model.f_pvalue))
```

독립변수 x = fixedacidity, 종속변수 y = pH로 두고 단순 선형 회귀 분석을 시행한 결과이다.  

회귀모형의 유의성 검정 결과를 살펴보면, 검정 통계량 (F-statistic)은 151.0이고 유의확률 (Prob(F-statistic))은 0에 근사함을 알 수 있다.  
따라서 유의수준 5%에서 회귀직선은 유의하다고 볼 수 있다.  
또한 결정계수 (Multiple R-squared)는 0.433으로써 전체 자료의 산포 중 약 43.3%가 회귀직선으로 설명이 가능한 것을 알 수 있다.

적합된 회귀직선은 pH = -0.103 * fixedacidity + 4.115이다. 
-0.103로 표현되는 회귀직선의 기울기의 통계적 유의성을 검정한 결과, t-통계량이 -12.286이고, 유의확률이 0에 근사하므로 유의하다.  
4.115로 표현되는 회귀직선의 절편의 통계적 유의성을 검정한 결과, t-통계량이 64.218이고, 유의확률이 0에 근사하므로 유의하다.  
따라서, 유의수준 5%에서 이 회귀직선은 유의하다고 말할 수 있다.

# 예제 2-3.

```python
t, p = stats.pearsonr(wine["pH"], wine["alcohol"])
print("Pearson's correlation coefficient : %s, p-value : %s" %(t, p))
```

scipy.stats의 pearsonr함수를 활용하여 와인의 pH와 알코올 양의 상관관계를 분석하였다.  
두 변수 간의 상관계수는 0.491로 나타난다.  
상관관계 유무의 검정 결과, 유의확률은 0에 근사하는 것을 확인 할 수 있다.  
따라서 유의수준 5%에서 두 변수 사이의 상관관계가 존재한다고 볼 수 있다.

# 예제 2-4.

```python
model = ols("fixedacidity ~ pH + alcohol + residualsugar", wine).fit()
print(model.summary())

intercept = model.params[0]
coefficient1 = model.params[1]
coefficient2 = model.params[2]
coefficient3 = model.params[3]

print("FixedAcidity = %s + %s * pH + %s * alcohol + %s * residualsugar + e" %(intercept, coefficient1, coefficient2, coefficient3))
print("검정통계량 : %s, 유의확률: %s " %(model.fvalue, model.f_pvalue))
```

독립변수 x1 = pH, x2 = alcohol, x3 = residualsugar, 종속변수 y = fixedacidity 로 두고 중회귀 분석을 시행한 결과이다.

회귀모형의 유의성 검정 결과를 살펴보면, 검정 통계량 (F-statistic)은 50.43이고 유의확률 (Prob(F-statistic))은 0에 근사함을 알 수 있다.  
따라서 유의수준 5%에서 회귀선은 유의하다고 볼 수 있다.  
또한 수정 결정계수 (Adjusted Multiple R-squared)는 0.427으로서 전체 자료의 산포 중 약 42.7%가 회귀선으로 설명이 가능한 것을 알 수 있다.

적합된 회귀식은 fixedacidity = 21.664 - 4.058 * pH - 0.062 * alcohol + 0.029 * residualsugar + e이다.  
21.664로 표현되는 회귀직선의 절편의 통계적 유의성을 검정한 결과, t-통계량이 18.819이고, 유의확률이 0에 근사하므로 유의하다.  
-4.058로 표현되는 pH의 상관계수의 통계적 유의성을 검정한 결과, t-통계량이 -10.297이고, 유의확률이 0에 근사하므로 유의하다.  
-0.062로 표현되는 alcohol의 상관계수의 통계적 유의성을 검정한 결과, t-통계량이 -0.827이고, 유의확률이 0.409이므로 유의하지 않다.  
0.029로 표현되는 residualsugar의 상관계수의 통계적 유의성을 검정한 결과, t-통계량이 0.649이고, 유의확률이 0.517이므로 유의하지 않다.  

따라서, 유의수준 5%에서 fixedacidity를 설명함에 있어서 pH가 유의한 변수라고 볼 수 있다.
