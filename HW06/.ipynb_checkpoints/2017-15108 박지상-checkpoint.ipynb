# 예제 1

```python
import math
from scipy.stats import norm

p1 = 54 / 150
p2 = 43 / 150
p = (54+43) / (150+150)

print("예제 1-(1).")

n = norm(loc = 0, scale=1)
z = (p1-p2) / math.sqrt(p*(1-p)*(1/150 + 1/150))

p = 1-n.cdf(z)
za = n.ppf(1-0.05)

print("검정통계량 = %s, 유의확률 = %s, 기각역: z >= %s" %(round(z, 5), round(p, 5), round(za, 5)))

if(p < 0.05 and z >= za):
    print("유의수준 5%에서 p1이 p2보다 높다는 유의한 증거가 있다.")
else:
    print("유의수준 5%에서 p1이 p2보다 높다는 유의한 증거가 없다.")

print("-" * 100)
print("예제 1-(2).")
upper = (p1-p2) + n.ppf(1-0.05/2) * math.sqrt(p1*(1-p1)/150 + p2*(1-p2)/150)
lower = (p1-p2) - n.ppf(1-0.05/2) * math.sqrt(p1*(1-p1)/150 + p2*(1-p2)/150)

print("p1-p2의 95% 신뢰구간: (", round(lower, 5), ",", round(upper, 5), ")")
```

## 결과해석

### 예제 1-(1).  
귀무가설: p1＝p2 이고 대립가설: p1 > p2 일 때,검정통계량은 1.35775이고 유의확률은 0.08727이다.  
유의수준 5%에서 기각역은 Z ≥ 1.64485이므로  
유의확률이 5%보다 크고, 검정통계량 z가 기각역에 포함되 않으므로 귀무가설을 기각할 수 없다.  
따라서 p1이 p2보다 크다는 유의한 증거가 없다.  

### 예제 1-(2).  
두 모비율의 차에 대한 구간추정에서 p1-p2의 95% 신뢰구간을 구하였을 때 (-0.0322, 0.17887)이 도출된다.

# 예제 2

```python
import pandas
import math
from scipy import stats

data1 = pandas.read_csv('./data1.csv')

print("예제 2-(1).")
print("귀무가설: 학년과 안경 착용 여부는 서로 독립이다.")
print("대립가설: 학년과 안경 착용 여부는 독립이 아니다.")

print("-"*100)

print("예제 2-(2).")
y = data1.groupby(['학년', '안경 유무']).size()
table = y.unstack()
print(table)

print("-"*100)
print("예제 2-(3).")
g, p, dof, expctd = stats.chi2_contingency(table)

print("Chi-squared test: X-squared = %s, p-value = %s, df = %s, expected value = \n %s" %(round(g, 5), round(p, 5), round(dof, 5), expctd))
```

## 결과해석

### 예제 2-(1).
학년과 안경 착용 사이에 연관이 있는지 검정하는 것은 두 속성의 독립성 검정이다.  
이 때, 귀무가설은 학년과 안경 착용 여부는 서로 독립이라는 것이고,  
대립가설은 학년과 안경 착용 여부는 독립이 아니라는 것이다.

### 예제 2-(2).
data1의 '학년'과 '안경 유뮤' 변수를 기준으로 분학표(관측도수 행렬)을 생성한다.

### 예제 2-(3).
독립성 검정의 검정통계량은 9.91808이고, 유의확률은 0.01928이다.  
따라서, 유의확률이 0.05보다 작으므로 유의수준 5%에서 학년과 안경 유무가 독립이라는 귀무가설을 기각할 수 있다.  
즉, 학년과 안경 착용 사이에 연관이 있다.
