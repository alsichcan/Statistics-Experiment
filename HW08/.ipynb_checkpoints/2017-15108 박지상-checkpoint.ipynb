# 예제 1.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import ProbPlot

data = pd.read_csv('통실_corona.csv')

data
```

# 예제 1-(1).

```python
num = data.values.reshape(828)
age = np.tile(range(0, 9), 92)
corona = pd.DataFrame({'num':num, 'age':age})

print(corona)

total = 0
for i in range(len(corona)):
    total += corona._get_value(i, 'num')

print(total)
```

주어진 데이터가 2020년 6월 1일 확진자부터 2020년 8월 31일 확진자의 수이므로,  
모든 행의 num 값을 합하였더니, 8512명이 도출되었다.

# 예제 1-(2).

```python
model = ols('num ~ C(age)', data = corona).fit()
table = sm.stats.anova_lm(model, typ = 2)
print(table)
```

일원배치법의 모형을 "확진자 수 = 총평균 + 나이대에 따른 처리 효과 + 오차항"으로 나타내었을 때, 

귀무가설: 나이대에 따른 처리 효과가 모두 0으로 동일하다.  
대립가설: 귀무가설이 옳지 않다.

분산분석 결과, 검정통계량 값은 15.0242이고 유의확률은 1.133984e-20으로 매우 작은 것으로 나타났다.  
따라서 유의수준 5%내에서 귀무가설을 기각할 수 있다.  

# 예제 1-(3).

```python
list_10 = np.where(corona['age']==1)[0]+1
list_20 = np.where(corona['age']==2)[0]+1

resid_10 = model.resid.iloc[list_10].tolist()
resid_20 = model.resid.iloc[list_20].tolist()

# 10대의 QQ-plot
QQ10 = ProbPlot(np.array(resid_10))
plot = QQ10.qqplot(line='s', color='C0', lw=1)
plt.title("10-19 Q-Q Plot")
plt.show()

# 20대의 QQ-plot
QQ20 = ProbPlot(np.array(resid_20))
plot = QQ20.qqplot(line='s', color="C0", lw=1)
plt.title("20-29 Q-Q Plot")
plt.show()
```

10대와 20대의 잔차에 대한 QQ-Plot을 그려보았을 때,  
자료값이 주어진 직선에 크게 밀접하게 분포되어 있지않고, 양 극단에서도 크게 벗어나는 모습을 보이므로  
정규성을 충족시키지 못한다고 할 수 있다.

# 예제 2.

```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import statsmodels.api as sm
from statsmodels.formula.api import ols

run10Samp = pd.read_csv('run10Samp.txt', sep = " ")
run10Samp
```

# 예제 2-(1).

```python
run = pd.DataFrame({'gender': run10Samp.gender, 'time': run10Samp.time})

model = ols('time ~ C(gender)', data = run).fit()
table = sm.stats.anova_lm(model, typ = 2)
print(table)
```

일원배치법의 모형을 "완주시간 = 총평균 + 성별에 따른 처리 효과 + 오차항"으로 나타내었을 때,

귀무가설: 성별에 따른 처리 효과가 모두 0으로 동일하다.
대립가설: 귀무가설이 옳지 않다.

분산분석 결과, 검정통계량 값은 26.186817이고 유의확률은 0.000002으로 매우 작은 것으로 나타났다.  
따라서 유의수준 5%내에서 귀무가설을 기각할 수 있다.

# 예제 2-(2).

```python
time_m = run10Samp[run10Samp['gender']=='M'].time.tolist()
time_f = run10Samp[run10Samp['gender']=='F'].time.tolist()

# 남성의 QQ-plot
QQm = ProbPlot(np.array(time_m))
plot = QQm.qqplot(line='s', color='C0', lw=1)
plt.title("Male Q-Q Plot")
plt.show()

# 여성의 QQ-plot
QQf = ProbPlot(np.array(time_f))
plot = QQf.qqplot(line='s', color="C0", lw=1)
plt.title("Female Q-Q Plot")
plt.show()
```

성별에 따른 완주시간의 분포에 대한 QQ-Plot을 그려보았을 때,  
남성과 여성 모두 양 극단에서 어느정도의 편차가 관찰되기는 하지만, 대체로 분포가 직선에 밀접하게 나타나기 때문에,  
완주시간의 분포가 정규분포를 따른다고 볼 수 있다.

# 예제 2-(3).

```python
ttest_ind(time_m, time_f, equal_var=True)
```

귀무가설 : 남성의 평균 완주시간과 여성의 평균 완주시간이 동일하다.  
대립가설 : 남성의 평균 완주시간과 여성의 평균 완주시간이 동일하지 않다.  

등분산을 가정한 두 모평균의 차이를 비교하였을 때, 검정통계량은 -5.1173이고 유의확률은 1.5438819783437654e-06로 매우 작은 것으로 나타났다.  
따라서 유의수준 5%에서 귀무가설을 기각할 수 있다.

# 예제 2-(4).

(1)에서 구한 검정통계량 값 (F값)은 26.186817이고, (3)에서 구한 검정통계량 값 (T값)은 -5.1173으로, F = T^2가 성립한다.  
이는 (3)에서의 검정통계량 T값은 자유도가 (남성 수 + 여성 수 - 2)인 T분포를 따르고,  
(1)에서의 검정통계량 F값은 자유도가 (1, 남성 수 + 여성수 -2)인 F분포를 따르기 때문이다. (성별 요인의 수준의 개수가 2개)

# 예제 2-(5).

```python
def changeAge(x):
    return x // 10

run10Samp['age'] = run10Samp.apply(lambda x: changeAge(x.age), axis = 1)
run10Samp
```

이원배치법의 모형을 "완주시간" = 총평균 + 성별에 따른 처리 효과 + 연령대에 따른 처리 효과 + 오차항"으로 나타내었을 때,  
각 요인에 대한 귀무가설과 대립가설은 다음과 같다.

1) 성별에 따른 처리 효과에 대한 가설 검정  
귀무가설 : 성별에 따른 처리 효과가 모두 0으로 동일하다.  
대립가설 : 귀무가설이 옳지 않다.  
    
2) 연령대에 따른 처리 효과에 대한 가설 검정  
귀무가설 : 연령대에 따른 처리 효과가 모두 0으로 동일하다.  
대립가설 : 귀무가설이 옳지 않다.

# 예제 2-(6).

```python
runGroup = pd.DataFrame({'age': run10Samp.age, 'gender': run10Samp.gender, 'time': run10Samp.time})
model = ols('time ~ C(age) + C(gender)', data = runGroup).fit()
table = sm.stats.anova_lm(model, typ=2)
print(table)
```

분산분석 결과

1) 성별에 따른 처리 효과에 대한 가설 검정에 있어서는  
검정통계량 값이 1.689469이고, 유의확률은 0.158923이므로, 유의수준 5%에서 귀무가설을 기각할 수 없다.  

2) 연령대에 따른 처리 효과에 대한 가설 검정에 있어서는  
검정통계량 값이 27.062973이고, 유의확률은 0.000001이므로, 유의수준 5%에서 귀무가설을 기각할 수 있다.  

# 예제 2-(7).

```python
seaborn.pointplot(x = "age", y="time", hue="gender", data=runGroup)
plt.show()
```

성별 요인과 연령대 요인에 대한 교호작용의 여부를 확인하기 위해 interaction plot을 그려보았다.  
평균 그림을 살펴보았을 때, 두 분포가 평행하게 분포하지 않는 것으로 보아 두 요인 간에 상호작용이 있다고 볼 수 있다.  
다만, 연령대에 따른 편차 정도가 대체로 일정하기 때문에, 상호작용이 크게 존재한다고 보기는 어렵다.
