# 예제 1.

```python
import numpy as np
np.random.seed(73)
alp = ["S", "O", "L", "V", "E"]

x = np.random.choice(alp, 36, replace = True)
x = np.reshape(x, (6, 6))

# 예제 1-1.
print("1-1.") 
print(x)
print('-' * 100)

# 예제 1-2.
print("1-2.")
print(x[0][3])
print('-' * 100)

# 예제 1-3.
print("1-3.")
print(x[:,[0, 2, 3]])
print('-' * 100)

# 예제 1-4.
print("1-4.")
crossArray = []
for i in range(len(x)):
    crossArray.append(x[i][i])
print(crossArray)
```

## 결과해석
1-1. alp 변수에 있는 5개 문자를 총 36개 랜덤복원추출한 뒤, (6,6) 행렬로 reshape한 모양이다.  
1-2. (6,6)행렬의 (1,4)값이 출력되었다.  
1-3. (6,6)행렬의 1, 3, 4열만을 추출해 출력하였다.  
1-4. (6,6)행렬의 주대각 성분, 즉 (1,1)부터 (6,6)까지의 문자로 이루어진 리스트가 출력되었다.

# 예제 2.

```python
import pandas as pd
df = pd.read_csv("cdc.txt", sep=" ")

# 예제 2-1. 1)
print("2-1. 1)")
print(df.iloc[19:25, 4:6])
print('-' * 100)

# 예제 2-1. 2)
print("2-1. 2)")
print(df.iloc[19:25, 7])
print('-' * 100)

# 예제 2-2.
print("2-2.")
print("Average height of female:", df.loc[df.gender == 'f'].height.mean())
print("Average height of male:", df.loc[df.gender == "m"].height.mean())
print('-' * 100)

# 예제 2-3.
print("2-3.")

maleFrame = df.loc[df.gender == 'm']
maleAge = maleFrame.age.mean()
print("Weight of male over average age")
print("Average:", maleFrame.loc[maleFrame.age >= maleAge].weight.mean(), "Variance:", maleFrame.loc[maleFrame.age >= maleAge].weight.var())

femaleFrame = df.loc[df.gender == 'f']
femaleAge = femaleFrame.age.mean()
print("Weight of female over average age")
print("Average:",  femaleFrame.loc[femaleFrame.age >= femaleAge].weight.mean(), "Variance:", femaleFrame.loc[femaleFrame.age >= femaleAge].weight.var())
```

## 결과해석
2-1.  
1) 20부터 25번째까지의 height, weight 값  
2) 20부터 25번째까지의 age값  

2-2. dataframe에서 각각 gender 항목이 'f'와 'm'로 일치하는 항목들에 대해 키의 평균 값을 구하였다.

2-3. dataframe에서 각각 gender 항목이 'm'과 'f'로 일치하는 maleFrame과 femaleFrame을 구한 뒤,   
각 Frame에서의 age값의 평균을 구하여 maleAge와 femaleAge로 설정한다.  
이후 maleFrame과 femaleFrame에서 나이가 평균보다 큰 항목들에 대해 weight 값의 평균과 분산을 구하였다.

# 예제 5.

```python
import numpy as np

# 예제 5-1.
print("5-1.")
base = np.array([[1], [2], [3], [4]])

arrayA = np.repeat(base, 2, axis = 0).flatten('C')
arrayB = np.repeat(base, 2, axis = 1).flatten('F')
print("A =", arrayA)
print("B =", arrayB)
print('-' * 100)

# 예제 5-2.
print("5-2.")
hstackArray = np.stack((arrayA, arrayB))
print("병합 행렬 예1:")
print(hstackArray)

vstackArray = np.stack((arrayA, arrayB), axis = -1)
print("병합 행렬 예2:")
print(vstackArray)
```

## 결과해석
5-1. 기본이 되는 행렬에 대해 repeat이 되는 축을 달리하여 A와 B를 만들었다.

5-2. A와 B에 대해서 하나는 행을 기준으로 병합하였고, 하나는 열을 기준으로 병합하였다.

# 예제 7.

```python
val = 10
while(True):
    if(val == 5):
        print("Good bye!")
        break
    print("Current variable value : ", val)
    val = val - 1
```

## 결과해석
val 값을 10으로 초기화한 뒤, while loop에서 val이 5가 된다면 Good bye!를 출력하고 loop을 종료하도록 break을 작성한다.  

val이 5가 아니라면 현재 값을 출력하고, 값을 1만큼 낮춘다.

# 예제 8.

```python
# 예제 8-1.
print("8-1.")
total1 = 0
for i in range(1, 101):
    if(i % 3 == 0 or i % 5 == 0):
        total1 += i
print("Total:", total1)
print('-' * 100)

# 예제 8-2.
print("8-2.")
def mul3or5(n):
    total2 = 0
    for i in range(1, n+1):
        if(i % 3 == 0 or i % 5 == 0):
            total2 += i
    return total2


if(mul3or5(100) == total1):
    print("8-1의 결과와 같다.")
else:
    print("8-1의 결과와 다르다.")
```

## 결과해석
8-1. 1부터 100까지(range(1,101))의 수에 대해서 차례대로 3의 배수 또는 5의 배수인지 판별하여  
조건에 맞을 경우 total1에 더한다.

8-2. 1부터 n까지 (range(1, n+1))의 수에 대해서 차례대로 3의 배수 또는 5의 배수인지 판별하여
조건에 맞을 경우 total2에 더한 다음 total2를 return하는 함수 mul3or5을 정의한다.
이후 total1과 mul3or5(100)의 리턴값을 비교하였고, 둘이 일치한다는 결과가 나왔다.

# 예제 9.

```python
# 예제 9-1.
print("9-1.")
i = 1
while(i <= 10):
    print(i)
    i += 1
print('-' * 100)

# 예제 9-2.
print("9-2.")
i = 0
while(i <= 9):
    i += 1
    if(i == 5): continue
    print(i)
    
```

## 결과해석
9-1. 초기값 i를 1로 설정해두고, i 값이 10보다 작거나 같은 동안 i값을 출력하고, i값을 1만큼씩 증가시킨다.
  
9-2. 초기값 i를 0으로 설정해두고, i 값이 9보다 작거나 같은 동안 i값을 1만큼씩 증가시키고 i 값을 출력한다.
i가 5일 때에는 continue를 통해 출력되지 않고 새롭게 loop이 시작된다.

# 예제 10.

```python
def BMI(weight, height): # weight (kg), height (cm)
    bmi = weight / (height/100)**2
    if(bmi < 18.5):
        print("마른체형")
    elif(bmi < 25.0):
        print("표준")
    elif(bmi < 30.0):
        print("비만")
    else:
        print("고도 비만")

BMI(63, 180)
BMI(91, 165)
```

## 결과해석
인수로 체중과 신장을 받는 함수 BMI에서 bmi를 먼저 계산한 뒤 (인수는 cm단위로 받기 때문에 m단위로 변환하는 작업을 거침)  
각 범위에 따라 결과값을 출력한다.
