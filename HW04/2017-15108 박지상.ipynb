# 예제 1

```python
import numpy as np

life = np.array([51, 52, 59, 57, 48, 57, 45, 58, 50, 42, 54, 43, 55, 55, 56, 56, 50, 58, 49, 52])

# 예제 1-(1)
print("1-(1).")
print("표본평균:", life.mean())
print("표본표준편차:", life.std())

print("-" * 100)
# 예제 1-(2)
print("1-(2).", end = ' ')

z = (life.mean() - 50)/(6/np.sqrt(20))
print("검정통계량:", z)
print("검정통계량이 기각역에 속하므로 귀무가설을 기각한다.")
```

## 결과해석

1-(1). 전구 20개의 수명에 대한 평균과 표준편차를 numpy method를 사용해 구하였다.  
1-(2). 전구 수명에 대한 검정통계량을 산출한 뒤, 유의수준 5%의 기각역 Z >= 1.645에 속하므로 귀무가설을 기각한다.

# 예제 2

```python
import numpy as np
from scipy.stats import norm
np.random.seed(15108)

# 예제 2-(1)
print("2-(1)")
x10 = norm.rvs(loc = 1, scale = 10, size = 10)
x1000 = norm.rvs(loc = 1, scale = 10, size = 1000)

print("mean of x10:", x10.mean())
print("mean of x1000:", x1000.mean())

print("-" * 100)
# 예제 2-(2)
print("2-(2)")

z10 = (x10.mean() - 0)/(10/np.sqrt(10))
z1000 = (x1000.mean() - 0)/(10/np.sqrt(1000))

print("x10을 표본으로 할 때: ", end = '')
if(z10 >= 1.645):
    print("귀무가설을 기각한다.")
else:
    print("귀무가설을 채택한다.")
    
print("x1000을 표본으로 할 때: ", end = '')
if(z1000 >= 1.645):
    print("귀무가설을 기각한다.")
else:
    print("귀무가설을 채택한다.")
    
x10_pVal = 1 - norm(0, 10).cdf((x10.mean()-0)/(10/np.sqrt(10))) 
x1000_pVal = 1 - norm(0, 1000).cdf((x1000.mean()-0)/(10/np.sqrt(1000)))

print("x10을 표본으로 할 때의 p-value:", x10_pVal)
print("x1000을 표본으로 할 때의 p-value:", x1000_pVal)
```

## 결과해석
2-(1). 학번 15108을 np.random.seed()로 입력하고, 평균 = 1, 표준편차 = 10인 정규분포에서 표본 10개, 1000개를 뽑아 각 변수에 저장한 뒤,  
       각 변수의 평균을 구한다.
  
2-(2). 표본의 갯수가 10일 때와 1000일 때의 검정통계량 값을 구하고, 유의 수준 5%에서의 기각역에 포함되는지에 따라 귀무가설의 채택/기각 여부를 판단한다. 이에 더해 p-Value 값을 구한다. 

# 예제 3

```python
import numpy as np
from scipy.stats import norm

np.random.seed(1)
mat = np.zeros(shape=(1000, 20))
alpha = 0.1
count = 0

for i in range(1000):
    mean = i+1
    sigma = i+1
    n1 = norm(loc = mean, scale = sigma)
    n2 = norm(loc = 0, scale = 1)
    mat[i,:] = n1.rvs(size = 20)
    upper = mat[i, :].mean() + n2.ppf(1-alpha/2) * (sigma/np.sqrt(20))
    lower = mat[i, :].mean() - n2.ppf(1-alpha/2) * (sigma/np.sqrt(20))
    
    if (lower < mean) & (mean < upper): 
        count = count + 1

print(count/1000)
```

## 결과해석
i = 1 ~ 1000에 대해 평균 = i, 표준편차 = i인 정규분포에서 20개의 표본을 추출한 뒤,  
90%의 신뢰구간을 구한 뒤, 각각에 대해 모평균 = i를 포함하고 있는 신뢰구간의 개수를 구하였다.  
이 결과 1000개 중 892개의 신뢰구간이 모평균을 포함하고 있다는 결과가 도출되었다.
